### Matplotlib
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

Package Homepage: https://matplotlib.org/  
Python Documentation (latest): https://matplotlib.org/stable/contents.html

We will use *matplotlib* to visualize data within our code.

#### Examples

```python
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4])
y = np.array([0, 2, 3, 2, 0])

# Matplotlib can be used with a global `plt.plot` `plt.scatter`, etc.
# This will create a figure with one axis. This is quick and dirty.
plt.plot(x, y)
# To modify the axis, you can use plt.gca() (GetCurrentAxis)
ax = plt.gca()
ax.set_title('A graph')
ax.set_xlabel('The x axis')
```

```python
# If you need more control, you can explicitly create the figure/axis objects
# This is useful if you need to plot more than one graph at once.
fig = plt.figure(figsize=(10, 3))
ax = plt.subplot(1, 1, 1) # uses most recently created figure object
ax.plot(x, y)
ax.set_title('Another graph')
ax.set_ylabel('The y axis')
```

### OpenCV
OpenCV (Open Source Computer Vision Library) is an open-source library that includes several hundreds of computer vision algorithms.

Package Homepage: https://opencv.org/  
Python Documentation (v4.5.2): https://docs.opencv.org/4.5.2/

We will use *OpenCV* to help us:  

- Load images
- Modify the colour space of images
- Apply Image Processing algorithms to images

#### Examples

```python
import cv2

# OpenCV loads data as a numpy array
file_path = './flower.jpg'
img = cv2.imread(file_path)
print(img.shape)
print(type(img))
print(img.dtype)

# We can modify the image with openCV functions
# Make image grey
img_grey = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
# Blur image
img_grey_blurred = cv2.blur(img_grey, (5, 5))

# NOT OPENCV: We use matplotlib to display images in the notebook
import matplotlib.pyplot as plt
plt.imshow(img_grey, cmap='gray', vmin=0, vmax=255)
plt.show()
plt.imshow(img_grey_blurred, cmap='gray', vmin=0, vmax=255)
plt.show()
```
